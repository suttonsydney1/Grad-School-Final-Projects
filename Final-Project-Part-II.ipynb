```python
import getpass
mypasswd = getpass.getpass()
username = 'ssdn4'
host = 'pgsql.dsa.lan'
database = 'dsa_student'
```

```python
from sqlalchemy.engine.url import URL
from sqlalchemy import create_engine

postgres_db = {'drivername': 'postgres',
               'username': username,
               'password': mypasswd,
               'host': host,
               'database' :database}
engine = create_engine(URL(**postgres_db), echo=True)
del mypasswd
```

```python
query = """


DROP TABLE IF EXISTS ssdn4.crime_location CASCADE;
CREATE TABLE ssdn4.crime_location(
    location_id INT GENERATED BY DEFAULT AS IDENTITY PRIMARY KEY,
    district INT,
    beat INT,
    block VARCHAR(50),
    ward INT,
    community_area INT,
    location_description VARCHAR(50),
    latitude INT,
    longitude INT
);

DROP TABLE IF EXISTS ssdn4.crime_code CASCADE;
CREATE TABLE ssdn4.crime_code(
    cc_id INT GENERATED BY DEFAULT AS IDENTITY PRIMARY KEY,
    iucr VARCHAR(50) UNIQUE,
    fbi_code VARCHAR(50) UNIQUE,
    primary_type VARCHAR(100),
    description VARCHAR(100)
);


DROP TABLE IF EXISTS ssdn4.crime_event;
CREATE TABLE ssdn4.crime_event(
    crime_id VARCHAR(50),
    case_number VARCHAR(50),
    arrest VARCHAR(50),
    domestic VARCHAR(50),
    date DATE,
    cc_id INT,
    location_id INT,
    updated_on DATE,
    FOREIGN KEY (cc_id) REFERENCES crime_code(cc_id),
    FOREIGN KEY (location_id) REFERENCES crime_location(location_id)
);
    
    
"""

with engine.connect() as connection:
    res = connection.execute(query)
    print(res)
```

## 4.1 Load the data from the given csv file to the relations



* Assuming there will be more than one relations, you need to extract a subsets of data from the csv data. As Python may not be your first choice, you can use any languages to create subsets of data. Then store these data into the M8 exercises folder.

```python
#create subsets of the data from original file

import pandas as pd
datapath = "/dsa/data/DSA-7030/Chicago-Crime-Sample-2012.csv"
df = pd.read_csv(datapath, index_col=0)
df.head()
```

```python
#Crime code dataframe

crime_code_df = df[['IUCR','FBI Code', 'Primary Type', 'Description']]
crime_code_df = crime_code_df.rename(columns={'IUCR':'iucr', 'FBI Code':'fbi_code',
                                             'Primary Type':'primary_type', 'Description':'description'})
crime_code_df1 = crime_code_df.drop_duplicates(subset = ['iucr'])
crime_code_df2 = crime_code_df1.drop_duplicates(subset =['fbi_code'])

```

```python
#crime location df

crime_location_df = df[['District', 'Beat', 'Block', 'Ward', 'Community Area', 'Location Description',
                       'Latitude', 'Longitude']]
crime_location_df = crime_location_df.rename(columns = {'District': 'district', 'Beat':'beat', 'Block':'block', 
                                             'Ward':'ward', 'Community Area':'community_area',
                                             'Location Description':'location_description', 'Latitude':'latitude',
                                             'Longitude':'longitude'})

crime_location_df1 = crime_location_df.drop_duplicates(subset=['district','beat','block','ward','community_area',
                                                              'location_description','latitude','longitude'])
crime_location_df1.head()
```

```python
#set unique index
crime_code_df2 = crime_code_df2.reset_index()
crime_code_df2 = crime_code_df2.rename(columns = {'index':'cc_id'})
crime_code_df2.head() 

crime_location_df1 = crime_location_df1.reset_index()
crime_location_df1 = crime_location_df1.rename(columns = {'index':'location_id'})
crime_location_df1.head()
```

```python
#merge info onto crime event data frame
int1 = df[['ID', 'Case Number', 'Arrest', 'Domestic', 'Date', 'Updated On', 'IUCR','Block']]
int1 = int1.rename(columns={'Block':'block','IUCR':'iucr'})
int1_df = crime_code_df2.merge(int1, on='iucr')
int2_df = int1_df.merge(crime_location_df1, on='block')
int2_df.head()
```

```python
crime_event_df = int2_df[['ID','Case Number', 'Arrest','Domestic', 'Date', 'Updated On', 'cc_id', 'location_id']]
crime_event_df = crime_event_df.rename(columns = {'ID':'crime_id', 'Case Number':'case_number', 'Arrest':'arrest',
                                                  'Domestic':'domestic', 'Date':'date','Updated On':'updated_on'})
crime_event_df = crime_event_df.drop_duplicates(subset=['crime_id','case_number','arrest','domestic','date',
                                                       'updated_on', 'cc_id', 'location_id'])
```

```python
crime_code_df2.to_sql('crime_code',
               engine,
               schema = username,
               if_exists = 'append',
               index=False,
               chunksize=1000)
```

```python
crime_location_df1.to_sql('crime_location',
               engine,
               schema = username,
               if_exists = 'append',
               index=False,
               chunksize=1000)
```

```python
crime_event_df.to_sql('crime_event',
               engine,
               schema = username,
               if_exists = 'append',
               index=False,
               chunksize=15000)
```
